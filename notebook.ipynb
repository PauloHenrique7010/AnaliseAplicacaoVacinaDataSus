```python
#Paulo Henrique da Silva
#Análise de imunizações nos ultimos 5 Anos, taxa de abandono, vacinas aplicadas, area de cobertura


#dados retirados de:
#https://datasus.saude.gov.br/acesso-a-informacao/imunizacoes-desde-1994/
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
cobertura15 = pd.read_excel('planilhas/taxa_cobertura2015.xlsx', skiprows = range(0, 4)) # pula as primeiras 5 linhas
cobertura16 = pd.read_excel('planilhas/taxa_cobertura2016.xlsx', skiprows = range(0, 4)) # pula as primeiras 5 linhas
cobertura17 = pd.read_excel('planilhas/taxa_cobertura2017.xlsx', skiprows = range(0, 4)) # pula as primeiras 5 linhas
cobertura18 = pd.read_excel('planilhas/taxa_cobertura2018.xlsx', skiprows = range(0, 4)) # pula as primeiras 5 linhas
cobertura19 = pd.read_excel('planilhas/taxa_cobertura2019.xlsx', skiprows = range(0, 4)) # pula as primeiras 5 linhas
cobertura20 = pd.read_excel('planilhas/taxa_cobertura2020.xlsx', skiprows = range(0, 4)) # pula as primeiras 5 linhas

cobertura15 = cobertura15.rename(columns={'Unidade da Federação': 'uf', 'Coberturas Vacinais': 'cobertura'})
cobertura16 = cobertura16.rename(columns={'Unidade da Federação': 'uf', 'Coberturas Vacinais': 'cobertura'})
cobertura17 = cobertura17.rename(columns={'Unidade da Federação': 'uf', 'Coberturas Vacinais': 'cobertura'})
cobertura18 = cobertura18.rename(columns={'Unidade da Federação': 'uf', 'Coberturas Vacinais': 'cobertura'})
cobertura19 = cobertura19.rename(columns={'Unidade da Federação': 'uf', 'Coberturas Vacinais': 'cobertura'})
cobertura20 = cobertura20.rename(columns={'Unidade da Federação': 'uf', 'Coberturas Vacinais': 'cobertura'})

cobertura15 = cobertura15.dropna()
cobertura16 = cobertura16.dropna()
cobertura17 = cobertura17.dropna()
cobertura18 = cobertura18.dropna()
cobertura19 = cobertura19.dropna()
cobertura20 = cobertura20.dropna()
```

```python
fig, ax = plt.subplots()

anos = ['2015','2016','2017','2018','2019','2020']

#historico = [cobertura15.cobertura.iloc[0], cobertura16.cobertura.iloc[0], cobertura17.cobertura.iloc[0]]


for x in range(len(cobertura15.cobertura)):
    historico = [cobertura15.cobertura.iloc[x], 
                 cobertura16.cobertura.iloc[x], 
                 cobertura17.cobertura.iloc[x], 
                 cobertura18.cobertura.iloc[x], 
                 cobertura19.cobertura.iloc[x], 
                 cobertura20.cobertura.iloc[x]] 
    ax.plot(anos, historico, label=cobertura15.uf.iloc[x])
ax.legend()
plt.xlabel('Anos')
plt.ylabel('Porcentagem de cobertura')

plt.show()
```
